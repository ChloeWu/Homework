```python
#####  第一題  #####
#履約價格為9150，以10/14之選擇權價格所畫的報酬圖

import numpy as np
import matplotlib.pyplot as plt

K = 9150
I = 500
St = np.arange(K-I, K+I)
Cp = 63
Pp = 45.5

LC = np.maximum(St-K, 0) - Cp
SC = -LC
LP = np.maximum(K-St, 0) - Pp
SP = -LP

plt.subplot(221)
plt.plot(St,LC)
plt.ylabel("profit") 
plt.title("Long Call") 
plt.xticks([8600,9800])

plt.subplot(222)
plt.plot(St,LP)
plt.title("Long Put") 
plt.xticks([8600,9800])

plt.subplot(223)
plt.plot(St,SC)
plt.xlabel("Price") 
plt.ylabel("profit") 
plt.title("Short Call") 
plt.xticks([8600,9000,9400,9800])

plt.subplot(224)
plt.plot(St,SP)
plt.xlabel("Price") 
plt.title("Short Put") 
plt.xticks([8600,9000,9400,9800])

plt.show() 
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,3)
ax1.plot(St,SC)
plt.xticks([8700,8900,9100,9300, 9500])
plt.xlabel("Price") 
plt.ylabel("profit") 
plt.title("Short Call")

ax2 = fig.add_subplot(2,2,1, sharex=ax1)
ax2.plot(St,LC)
plt.ylabel("profit") 
plt.title("Long Call") 

ax3 = fig.add_subplot(2,2,4, sharey=ax1)
ax3.plot(St,SP)
plt.xticks([8700,8900,9100,9300, 9500])
plt.xlabel("Price") 
plt.title("Short Put")
 
ax4 = fig.add_subplot(2,2,2, sharex=ax3, sharey=ax2)
ax4.plot(St,LP)
plt.title("Long Put") 
 
plt.setp(ax2.get_xticklabels(), visible=False)
plt.setp(ax4.get_xticklabels(), visible=False)
plt.setp(ax4.get_yticklabels(), visible=False)
plt.setp(ax3.get_yticklabels(), visible=False)
 
plt.show()
```

```python
#####  第二題  #####
#Top strangle，以10/14之選擇權價格為例

Kc = 9250
Kp = 9000
C = 21
P = 10.5
St = np.arange(K-I, K+I)

SC1 = C - np.maximum(St-Kc, 0)
SP1 = P - np.maximum(Kp-St, 0)
top = SC1 + SP1

plt.plot(St, SC1, "b--")
plt.plot(St, SP1, "b--")
plt.plot(St, top, "r", lw = 3)
plt.xlabel("Price") 
plt.ylabel("profit") 
plt.title("Top strangle") 
plt.xlim(8900,9400)
plt.ylim(-40,50)

plt.show()
```
